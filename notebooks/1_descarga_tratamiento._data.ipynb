```python
##no hay columna  año en el DataFrame no_comparativo asi que lo vamos a extraer y convertir a int para que coincida con el otr CVS
df_no_comparativa['ano_evento'] = pd.to_datetime(df_no_comparativa["fecha_evento"]).dt.year
```

```python
df_no_comparativa.ano_evento.unique()
```

```python
###tambien en la columna hora2 hay un error de horas inexistentes como 80000 hrs
## por lo que extraeremos  la hora2 de la columna hora
df_no_comparativa['hora2'] = pd.to_datetime(df_no_comparativa['hora'], format='%H:%M:%S').dt.hour
```

#### Extrar columnas que se comparten en ambos CSV

```python
##checamos las columnas compartidas que  
columns_comparativa=list(df_comparativa.columns)
columns_no_comparativa=list(df_no_comparativa.columns)
columnas_compartidas=[x for x in columns_comparativa if x in columns_no_comparativa]
```

```python
# seleccionamos columnas comunes entre ambos DataFrames
df_comparativa_s=df_comparativa[columnas_compartidas]
df_no_comparativa_s=df_no_comparativa[columnas_compartidas]
##comprobamos que ahora si tengan el mismo numero de columnas
print(df_comparativa_s.shape)
print(df_no_comparativa_s.shape)
```

### Eliminamos los años 2018-2019 de la tabla no comparativa y unimos 

```python
###eliminamos los años 2018 y 2019 de la DF no comparativa por que son los que tienen errores
## lo anterior para para hacer un merge de las dos tablas 
df_no_comparativa_s=df_no_comparativa_s[df_no_comparativa_s["ano_evento"] > 2019]
df_no_comparativa_s.ano_evento.unique()
```

```python
## concatenamos
transito_concat=pd.concat([df_comparativa_s,df_no_comparativa_s], ignore_index=True)
```

```python
transito_concat
```

```python
#corroboramos que coinciden con la suma de las DataFrames originales 
print(transito_concat.shape[0])
print(df_no_comparativa_s.shape[0] + df_comparativa_s.shape[0])
```

### Seleccion de variables 

```python
transito_concat_a=transito_concat[["fecha_evento",	"ano_evento","mes","dia","hora2","condicion","tipo_de_evento","colonia","alcaldia","tipo_de_interseccion","tipo_de_vehiculo_1","tipo_de_vehiculo_2",
"lesiones","edad_lesionado","identidad","unidad_medica_de_apoyo","total_lesionados","total_occisos"]]
```

#### Es necesario corregir las columna 2, 3 y 9 

```python
### corregir el mes ya que algunos estan mayusculas y otros no 
transito_concat_a['mes']=transito_concat_a['mes'].apply(str.lower) 
```

```python
# En el dia hay unos en mayusculas y otros no, y ademas hay valores con erroes 
transito_concat_a['dia']=transito_concat_a['dia'].apply(str.lower)
```

```python
##checamos los errores 
transito_concat_a["dia"].unique()
```

```python
#remplazamos  
transito_concat_a=transito_concat_a.replace(to_replace=["miércoles",'miérco', 'mábado','sábado'],value=['miercoles','miercoles',"sabado","sabado"])
```

```python
## hay filas con otros valores que no son dias 
## En este caso repmplazaremos eso valors con la moda   
transito_concat_a["dia"].mode()[0]
```

```python
transito_concat_a=transito_concat_a.replace(['coruña', 'siérco', 'taxqueñes', 'calz taxqueña',
       'ruben leñero', 'taxqueña', 's<c3>érco', '1° de mayo'],"sabado")
```

```python
# checamos
transito_concat_a["dia"].unique()
```

```python
## Ahora vamos con el tipo de interseccion 
#en estre caso utilizaremos mayusculas para homologar
transito_concat_a.tipo_de_interseccion.unique()
```

```python
##remlazamos minusculas por mayusculas 
transito_concat_a.replace(to_replace=['y', 't'],value=['T', 'Y'], inplace=True)
```

```python
##checamos
transito_concat_a.tipo_de_interseccion.unique()
```
